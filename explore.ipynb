```python
import pandas as pd
import numpy as np
```

```python
_2012 = pd.read_excel(
    r"C:\Users\HP\Documents\Lab\On_It\Elections-Dashboard\data\2012 NDC WA CENTRAL COLLATION OF RESULT.xlsx",
    sheet_name=None,
)
sheet_names = list(_2012.keys())
first_sheet_name = sheet_names[0]
second_sheet_name = sheet_names[1]
presidential2012 = _2012[first_sheet_name]
parliamentary2012 = _2012[second_sheet_name]

presidential2012.columns = presidential2012.iloc[8]
presidential2012 = presidential2012.iloc[9:].copy()
presidential2012.drop(["NO.", "VVC", " % TO"], axis=1, inplace=True)
presidential2012.rename(columns={'P. STATION CODE': 'PS CODE'}, inplace=True)
presidential2012.rename(columns={'POLLING STATION': 'PS NAME'}, inplace=True)
presidential2012.rename(columns={'RB': 'REJECTED'}, inplace=True)
presidential2012.rename(columns={'TVC': 'TOTAL VOTES'}, inplace=True)
presidential2012["YEAR"] = 2012

parliamentary2012.columns = parliamentary2012.iloc[9]
parliamentary2012 = parliamentary2012.iloc[10:].copy()
parliamentary2012.drop(["NO.", "VVC", " % TO"], axis=1, inplace=True)
parliamentary2012.rename(columns={'P. STATION CODE': 'PS CODE'}, inplace=True)
parliamentary2012.rename(columns={'POLLING STATION': 'PS NAME'}, inplace=True)
parliamentary2012.rename(columns={'RB': 'REJECTED'}, inplace=True)
parliamentary2012.rename(columns={'TVC': 'TOTAL VOTES'}, inplace=True)
parliamentary2012["YEAR"] = 2012
```

```python
_2016 = pd.read_excel(
    r"C:\Users\HP\Documents\Lab\On_It\Elections-Dashboard\data\2016 election collation WA CENTRAL.xls",
    sheet_name=None,
)
sheet_names = list(_2016.keys())
first_sheet_name = sheet_names[0]
second_sheet_name = sheet_names[1]
presidential2016 = _2016[first_sheet_name]
parliamentary2016 = _2016[second_sheet_name]

presidential2016.columns = presidential2016.iloc[8]
presidential2016 = presidential2016.iloc[9:].copy()
presidential2016.drop(["NO.", "VVC", " % TO"], axis=1, inplace=True)
presidential2016.rename(columns={'Code': 'PS CODE'}, inplace=True)
presidential2016.rename(columns={'Name': 'PS NAME'}, inplace=True)
presidential2016.rename(columns={'RB': 'REJECTED'}, inplace=True)
presidential2016.rename(columns={'TVC': 'TOTAL VOTES'}, inplace=True)
presidential2016.rename(columns={'IND 1': 'IND'}, inplace=True)
presidential2016["YEAR"] = 2016

parliamentary2016.columns = parliamentary2016.iloc[8]
parliamentary2016 = parliamentary2016.iloc[9:].copy()
parliamentary2016.drop(["NO.", "VVC", " % TO"], axis=1, inplace=True)
parliamentary2016.dropna(axis=1, how="all", inplace=True)
parliamentary2016.rename(columns={'Code': 'PS CODE'}, inplace=True)
parliamentary2016.rename(columns={'Name': 'PS NAME'}, inplace=True)
parliamentary2016.rename(columns={'RB': 'REJECTED'}, inplace=True)
parliamentary2016.rename(columns={'TVC': 'TOTAL VOTES'}, inplace=True)
parliamentary2016.rename(columns={'IND 1': 'IND'}, inplace=True)

parliamentary2016["YEAR"] = 2016
```

```python
_2020 = pd.read_excel(
    r"C:\Users\HP\Documents\Lab\On_It\Elections-Dashboard\data\FINAL_COLLATION SHEET_2020_FINAL WA CENTRAL.xls",
    sheet_name=None,
)
sheet_names = list(_2020.keys())
first_sheet_name = sheet_names[0]
second_sheet_name = sheet_names[1]
presidential2020 = _2020[first_sheet_name]
parliamentary2020 = _2020[second_sheet_name]

presidential2020.columns = presidential2020.iloc[5]
presidential2020 = presidential2020.iloc[6:].copy()
presidential2020.drop(["S/N"], axis=1, inplace=True)
presidential2020.rename(columns={'Code': 'PS CODE'}, inplace=True)
presidential2020.rename(columns={'Name': 'PS NAME'}, inplace=True)
presidential2020["YEAR"] = 2020

parliamentary2020.columns = parliamentary2020.iloc[5]
parliamentary2020 = parliamentary2020.iloc[6:].copy()
parliamentary2020.drop(["S/N"], axis=1, inplace=True)
parliamentary2020.dropna(axis=1, how="all", inplace=True)
parliamentary2020.rename(columns={'Code': 'PS CODE'}, inplace=True)
parliamentary2020.rename(columns={'Name': 'PS NAME'}, inplace=True)
parliamentary2020["YEAR"] = 2020
```
